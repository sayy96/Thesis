```python
import matplotlib.pyplot as plt

# Function to draw the letter 'L'
def draw_L(ax, start_x, start_y):
    ax.plot([start_x, start_x], [start_y, start_y + 4], color='blue', lw=5)  # Vertical line
    ax.plot([start_x, start_x + 2], [start_y, start_y], color='blue', lw=5)  # Horizontal base

# Function to draw the letter 'E'
def draw_E(ax, start_x, start_y):
    ax.plot([start_x, start_x], [start_y, start_y + 4], color='green', lw=5)  # Vertical line
    ax.plot([start_x, start_x + 2], [start_y + 4, start_y + 4], color='green', lw=5)  # Top horizontal
    ax.plot([start_x, start_x + 1.5], [start_y + 2, start_y + 2], color='green', lw=5)  # Middle horizontal
    ax.plot([start_x, start_x + 2], [start_y, start_y], color='green', lw=5)  # Bottom horizontal

# Function to draw the letter 'Y'
def draw_Y(ax, start_x, start_y):
    ax.plot([start_x, start_x + 1], [start_y, start_y + 2], color='red', lw=5)  # Left diagonal
    ax.plot([start_x + 1, start_x + 2], [start_y + 2, start_y], color='red', lw=5)  # Right diagonal
    ax.plot([start_x + 1, start_x + 1], [start_y + 2, start_y + 4], color='red', lw=5)  # Vertical line

# Set up the figure and axis
fig, ax = plt.subplots()
ax.set_aspect('equal')

# Drawing letters "L", "E", and "Y"
draw_L(ax, 0, 0)    # Draw 'L'
draw_E(ax, 3, 0)    # Draw 'E'
draw_Y(ax, 6, 0)    # Draw 'Y'

# Formatting the plot
ax.set_xlim(-1, 9)
ax.set_ylim(-1, 6)
ax.axis('off')  # Hide axes

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Function to draw the letter 'L'
def draw_L(ax, start_x, start_y):
    ax.plot([start_x, start_x], [start_y, start_y + 4], color='blue', lw=5)  # Vertical line
    ax.plot([start_x, start_x + 2], [start_y, start_y], color='blue', lw=5)  # Horizontal base

# Function to draw the letter 'E'
def draw_E(ax, start_x, start_y):
    ax.plot([start_x, start_x], [start_y, start_y + 4], color='green', lw=5)  # Vertical line
    ax.plot([start_x, start_x + 2], [start_y + 4, start_y + 4], color='green', lw=5)  # Top horizontal
    ax.plot([start_x, start_x + 1.5], [start_y + 2, start_y + 2], color='green', lw=5)  # Middle horizontal
    ax.plot([start_x, start_x + 2], [start_y, start_y], color='green', lw=5)  # Bottom horizontal

# Function to draw the inverted 'Y'
def draw_inverted_Y(ax, start_x, start_y):
    ax.plot([start_x, start_x + 1], [start_y + 4, start_y + 2], color='red', lw=5)  # Left diagonal, inverted
    ax.plot([start_x + 1, start_x + 2], [start_y + 2, start_y + 4], color='red', lw=5)  # Right diagonal, inverted
    ax.plot([start_x + 1, start_x + 1], [start_y, start_y + 2], color='red', lw=5)  # Vertical line (bottom part)

# Set up the figure and axis
fig, ax = plt.subplots()
ax.set_aspect('equal')

# Drawing letters "L", "E", and the inverted "Y"
draw_L(ax, 0, 0)    # Draw 'L'
draw_E(ax, 3, 0)    # Draw 'E'
draw_inverted_Y(ax, 6, 0)  # Draw inverted 'Y'

# Formatting the plot
ax.set_xlim(-1, 9)
ax.set_ylim(-1, 6)
ax.axis('off')  # Hide axes

# Show the plot
plt.show()
```
